# Vamos começar com a plotar gráficos

## 1. Scatter plot simples (Dispersão)
Vamos começar com o plot padrão do matplotlib

```python
#Vamos começar com uma função do terceiro grau.
#Os argumentos são opcionais, por isso o default é 1

def func_terc_grau (x, a = 1, b = 1, c = 1, d = 1):
    y = a*x**3 + b*x**2 + c*x + d
    return y
```

```python
lista_x = range(-5, 5)
lista_y3 = []

for i in lista_x:
    lista_y3.append(func_terc_grau(i))
```

```python
plt.plot(lista_x, lista_y3)
plt.show()
```

```python
#Vamos começar a incrementar nosso gráfico

plt.plot(lista_x, lista_y3, label = 'f(x)', color = 'black')
plt.xlabel('Valores de X')
plt.ylabel('Valores de Y')
plt.legend()
plt.title('Função do Terceiro Grau')
plt.scatter(lista_x, lista_y3, color = 'blue')
plt.show()
```

```python
#Como colocar dois gráficos num mesmo comando?
#Vamos começar criando uma função do segundo grau também

def func_seg_grau (x, a = 1, b = 1, c = 1):
    y = a*x**2 + b*x + c
    return y
```

```python
lista_y2 = []

for i in lista_x:
    lista_y2.append(func_seg_grau(i))
```

```python
plt.plot(lista_x, lista_y2)
plt.title('Função do Segundo Grau')
plt.show()
```

```python
#Vamos plotar ambos os gráficos no mesmo plot:

plt.plot(lista_x, lista_y3, label = 'f(x)', color = 'black')
plt.xlabel('Valores de X')
plt.ylabel('Valores de Y')
plt.legend()
plt.title('Função do Segundo e Terceiro Grau')
plt.scatter(lista_x, lista_y2, color = 'blue')
plt.plot(lista_x, lista_y2)
plt.show()
```

## 2. Gráfico de Pizza

Também chamado de Pie Chart

Ideal pra entender distribuição

```python
#Vamos ver quantos pasteis determinada barraca vendeu durante a feita.

sabores = 'Carne Seca', 'Frango', 'Queijo', 'Pizza', 'Pernil', 'Carne Moída', 'Presunto e queijo' #Relação de sabores
vendas = [15, 10, 10, 4, 5, 9, 9] #Relação de vendas
menores = (0, 0, 0, 0.1, 0.1, 0, 0) #Como pizza e pernil venderam menos, vamos destacar para facilitar a visualização

plt.pie(vendas, explode = menores, labels = sabores)
plt.title('Pastelaria do Seu Zé')
plt.show()
```

```python
#Vamos fazer um gráfico a partir de um dicionário
#Vamos quantas horas cada pessoa aplicou no desenvolvimento de determinado projeto

grupo = {'Eliseu': 120, 'Mônica': 280, 'João': 85, 'Leonardo': 95, 'Maria': 150}
nomes = tuple(grupo.keys())
valores = list(grupo.values())
#Desmembramos o dicionário para acessar seus valores e chaves

plt.pie(valores, labels = nomes)
plt.title('Projeto XPTO')
plt.show()
```

## 3. Histograma

É um ótimo gráfico para entender a distribuição de valores

```python
#Vamos criar uma um array com 1 milhão de valores randômicos
N = 1000000
valores = np.random.randn(N)
```

```python
plt.hist(valores, bins = 20, color = 'orange')
plt.show()
```

A curva possui formato de sino, ou seja, a distribuição é normal.

Espera-se que 95% dos dados estejam distribuidos entre 2 e -2.

```python
#O argumento bins refere-se a quantidade de barras do histograma. Vamos mexer nisso

plt.hist(valores, bins = 100) #Com 100 barras, a curva fica mais suave
plt.show()
```
